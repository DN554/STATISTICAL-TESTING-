**T TEST**- Compare the means of two groups or to compare a sample mean to a known population mean when the population standard deviation is unknown and the sample size is relatively small

```python
from scipy import stats
import numpy as np
```

```python
sample_data=np.array([10,12,11,13,10,9,12,11,10,13])
```

```python
sample_data.mean()
```

```python
#ONE SAMPLE TEST
sample_data=np.array([10,12,11,13,10,9,12,11,10,13])
population_mean=10.5
t_statistic,p_value=stats.ttest_1samp(sample_data,population_mean)
print(f"one -sample t -test:t-statistic ={t_statistic:.2f},p-value={p_value:.3f}")
```

```python
#TWO SAMPLE INDEPENDET T-TEST
df1= np.array([25,28,27,26,29])
df2= np.array([22,24,23,21,25])
t_statistic,p_value=stats.ttest_ind(df1,df2)
print(f"Two sampke independent t-test:t-statistic={t_statistic:.2f},p-value={p_value:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#GENERATE EXAMPLE DATA FROM TWO GROUPS
np.random.seed(42)
group1=np.random.normal(loc=0,scale=1,size=100)
group2=np.random.normal(loc=0.5,scale=1,size=100)
```

```python
group1
```

```python
group2
```

```python
#PERFORM INDEPENDENT SAMPLE T-TEST
t_statistic,p_value=stats.ttest_ind(group1,group2,equal_var=True)
print(f"Two sample independent t-test:t-statistic={t_statistic:.2f},p-value={p_value:.3f}")
```

```python
#CREATE A FIGURE AND AXES FOR THE PLOT
plt.figure(figsize=(8,6))
```

```python
#PLOT HISTOGRAM OF THE TWO GROUP
sns.histplot(group1,label='Group 1',kde=True,alpha=0.6,stat="density",color="skyblue")
sns.histplot(group2,label='Group 2',kde=True,alpha=0.6,stat="density",color="salmon")
```

```python
#ADDING VERTICAL LINES FOR THE MEANS OF EACH GROUP
plt.figure(figsize=(5,4))
plt.axvline(np.mean(group1),color='skyblue',linestyle='--',linewidth=2,label='Mean Group 1')
plt.axvline(np.mean(group2),color='pink',linestyle='--',linewidth=2,label='Mean Group 2')
```

```python
plt.figure(figsize=(8, 6))
plt.plot(group1, label='Group 1', marker='o')
plt.plot(group2, label='Group 2', marker='o')
plt.title('Comparison of Two Groups')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

#ADD TEST ANNOTATIONS FOR T-STATISTIC AND P-VALUE
plt.text(0.05, 0.95, f't-statistic={t_statistic:.2f}', transform=plt.gca().transAxes, fontsize=12, color='black', verticalalignment='top')
plt.text(0.05, 0.90, f'p-value={p_value:.3f}', transform=plt.gca().transAxes, fontsize=12, color='black', verticalalignment='top')
```
